```python
df_america = poll.loc[poll['Country'].isin(['Argentina','United States of America'])]
df_europe = poll.loc[~poll['Country'].isin(['Argentina','United States of America'])]
```

```python
def dfs_tabs(df_list, sheet_list, file_name):
    writer = pd.ExcelWriter(file_name,engine='xlsxwriter')   
    for dataframe, sheet in zip(df_list, sheet_list):
        dataframe.to_excel(writer, sheet_name=sheet, startrow=0 , startcol=0)   
    writer.save()

# list of dataframes and sheet names
dfs = [poll, df_america, df_europe]
sheets = ['poll','america','europe']    

# run function
dfs_tabs(dfs, sheets, 'tableau.xlsx')
```
